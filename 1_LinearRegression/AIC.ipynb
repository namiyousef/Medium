# Akaike's information criterion

# Equation

$$AIC = 2\ln\left (\frac{e^k}{\hat{\mathcal{L}}}\right)$$

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

plt.style.use('classic')
```
