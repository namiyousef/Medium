# MAE, MSE, RMSE

## MAE Equation

$$ MAE = \frac{1}{n}\sum_{i=1}^{n}\left |y_i - \hat{y}_i \right |$$

## MSE Equation

$$ MSE = \frac{1}{n}\sum_{i=1}^{n}\left (y_i - \hat{y}_i \right )^2$$

## RMSE Equation

$$ RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}\left (y_i - \hat{y}_i \right )^2}$$

```python
# libraries

import numpy as np
import random

import matplotlib.pyplot as plt

plt.style.use('ggplot')

from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import mean_absolute_error as MAE
```

```python
def function(x, a, b):
    """
    takes an np.array, x, and outputs an np.array, y = a*x + b
    """
    return a*x + b
```

```python
x = np.linspace(0, 100, 101)

y = function(x, 2, 5)

y_noise = y + np.random.normal(0,5,101)

plt.plot(x,y, label = 'predicted data')
plt.plot(x,y_noise,'.',label = 'actual data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

print("MAE = {}".format(MAE(y_noise, y)))
print("MSE = {}".format(MSE(y_noise, y)))
print("RMSE = {}".format(np.sqrt(MSE(y_noise, y))))
```

```python
from statistics import stdev
x = np.linspace(0, 100, 10001)

y = function(x, 2, 5)

y_noise = y + np.random.normal(0,10,10001)

MAEs = []
RMSEs = []
length = []
for i in range(10000):
    LHS = random.randint(0,1000)
    RHS = random.randint(LHS+1,1001)
    MAEs.append(
        MAE(y_noise[LHS:RHS], y[LHS:RHS])
    )
    RMSEs.append(
        np.sqrt(MSE(y_noise[LHS:RHS], y[LHS:RHS]))
    )
    length.append(RHS-LHS)
    
plt.hist(RMSEs)
plt.hist(MAEs, alpha = 0.5)
plt.legend([stdev(RMSEs), stdev(MAEs)])
#plt.hist(MAEs)
#plt.show()

#plt.hist(RMSEs)
#plt.show()

plt.show()

plt.plot(length, RMSEs,'.',label= 'RMSE')
plt.plot(length, MAEs,'.', label = 'MAE')
plt.xlabel('length of dataset')
plt.ylabel('score')
plt.legend()
```
