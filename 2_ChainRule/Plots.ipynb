# Creating Plots for Chain Rule Project

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
food = np.linspace(0,10)
Hunger = -food**2 + 10*food

fig, ax = plt.subplots()
plt.plot(food, Hunger)
plt.ylim([0, 30])
plt.xticks([2,5,8])

labels = [item.get_text() for item in ax.get_xticklabels()]
labels[1] = 'Optimum food'
"""label = ax.set_xlabel('No food', fontsize = 9)
ax.xaxis.set_label_coords(0.1, -0.020)

label = ax.set_xlabel('Too much food', fontsize = 9)
ax.xaxis.set_label_coords(0.9, -0.020)
"""
ax.set_xticklabels(labels)
#plt.yticks([])
print(labels)

plt.text(0.1, -0.035, 'No food', horizontalalignment='center',verticalalignment='center', transform=ax.transAxes, color = 'gray')
plt.text(0.9, -0.035, 'Too much food', horizontalalignment='center',verticalalignment='center', transform=ax.transAxes, color = 'gray')

plt.yticks([])

```
