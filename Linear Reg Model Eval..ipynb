## Linear Regression Evaluation Metrics

```python
#import modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from scipy.optimize import curve_fit

#necessary functions
def LinearFunction(x,b1,b2):
    model = b1+ b2*x
    return model

```

```python
np.random.seed(1) #sets the seed
x = np.linspace(1,100,100) #create 100 points

#plots that do not deviate much
y = x
y_noise = y + np.random.normal(0,10,100)
model = y
plt.plot(x,y_noise,'o')
plt.plot(x,model,'r')
plt.title('Example of a good model with a strong fit')
plt.xlabel('x')
plt.ylabel('y_noise')
plt.xlim(0,100)
plt.ylim(-100,200)
plt.legend(['Actual Data','Fitted Data'])
plt.show()
print("R^2 = ",r2_score(y_noise,model))

x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
y_noise = y + np.random.normal(0,0,100)
plt.plot(x,y_noise,'o')


popt,pcov = curve_fit(LinearFunction, x,y_noise)


model = LinearFunction(x,*popt)
plt.plot(x,model,'r')
plt.title('Example of a bad model with a moderate fit')
plt.xlabel('x')
plt.ylabel('y_noise')
plt.legend(['Actual Data','Fitted Data'])
plt.show()
print("R^2 = ",r2_score(y_noise,model))



x = np.linspace(0,100,100)
y = x**2
y_noise = y + np.random.normal(0,0,100)
plt.plot(x,y_noise,'o')


popt,pcov = curve_fit(LinearFunction, x,y_noise)


model = LinearFunction(x,*popt)
plt.plot(x,model,'r')
plt.title('Example of a bad model with a strong fit')
plt.xlabel('x')
plt.ylabel('y_noise')
plt.legend(['Actual Data','Fitted Data'])
plt.show()
print("R^2 = ", r2_score(y_noise,model))


y = x
y_noise = y + np.random.normal(0,50,100)
model = y
plt.plot(x,y_noise,'o')
plt.plot(x,model,'r')
plt.title('Example of a good model with a weak fit')
plt.xlabel('x')
plt.ylabel('y_noise')
plt.xlim(0,100)
plt.ylim(-100,200)
plt.legend(['Actual Data','Fitted Data'])
plt.show()
print("R^2 = ",r2_score(y_noise,model))


y = x
y_noise = y + np.random.normal(0,10,100)
model = 1.25*x -25
plt.plot(x,y_noise,'o')
plt.plot(x,model,'r')
plt.title('Example of a bad model with a moderate fit')
plt.xlabel('x')
plt.ylabel('y_noise')
plt.xlim(0,100)
plt.ylim(-100,200)
plt.legend(['Actual Data','Fitted Data'])
plt.show()
print("R^2 = ",r2_score(y_noise,model))
```

```python
def Polynomial(x,b1,b2,b3,b4,b5,b6,b7,b8):
    model = b1+ b2*x+b3*x**2+b4*x**3+b5*x**4+b6*x**5+b7*x**6+b8*x**7
    return model

x = np.linspace(0,100,50)
y = x
y_noise = y + np.random.normal(0,20,50)
plt.plot(x,y_noise,'o')

popt,pcov = curve_fit(Polynomial, x,y_noise)
model = Polynomial(x,*popt)

plt.plot(x,model,'r')
plt.title('Polynomial, Data set 1')
plt.xlabel('x')
plt.ylabel('y_noise')
plt.xlim(0,100)
plt.ylim(-100,200)
plt.legend(['Actual Data','Fitted Data'])
plt.show()
print("R^2 = ",r2_score(y_noise,model))
print("R^2_Adj = ", 1-((1-r2_score(y_noise,model))*(len(y_noise)-1))/(len(y_noise)-7-1))

plt.plot(x,y_noise,'o')

model2 = y;

plt.plot(x,model2,'r')
plt.title('Linear, Data set 1')
plt.xlabel('x')
plt.ylabel('y_noise')
plt.xlim(0,100)
plt.ylim(-100,200)
plt.legend(['Actual Data','Fitted Data'])
plt.show()
print("R^2 = ",r2_score(y_noise,model2))
print("R^2_Adj = ", 1-((1-r2_score(y_noise,model2))*(len(y_noise)-1))/(len(y_noise)-1-1))


y_noise = y + np.random.normal(0,20,50)
plt.plot(x,y_noise,'o')
plt.plot(x,model,'r')
plt.title('Polynomial, Data set 2 (Out of Sample)')
plt.xlabel('x')
plt.ylabel('y_noise')
plt.xlim(0,100)
plt.ylim(-100,200)
plt.legend(['Actual Data','Fitted Data'])
plt.show()
print("R^2 = ",r2_score(y_noise,model))
print("R^2_Adj = ", 1-((1-r2_score(y_noise,model))*(len(y_noise)-1))/(len(y_noise)-7-1))

plt.plot(x,y_noise,'o')
plt.plot(x,model2,'r')
plt.title('Linear, Data set 2 (Out of Sample)')
plt.xlabel('x')
plt.ylabel('y_noise')
plt.xlim(0,100)
plt.ylim(-100,200)
plt.legend(['Actual Data','Fitted Data'])
plt.show()
print("R^2 = ",r2_score(y_noise,model2))
print("R^2_Adj = ", 1-((1-r2_score(y_noise,model2))*(len(y_noise)-1))/(len(y_noise)-1-1))

```
